```python
import pandas as pd
import requests
import json
from pandas.io.json import json_normalize
```

```python
table = pd.read_excel("https://files.cpicpgx.org/data/report/current/pair/cpic_gene-drug_pairs.xlsx")
table
```

```python
drugs = table["Drug"].tolist()
```

```python
drug_dict = {
    "format": "minimal",
    "terms": drugs
    
}
```

```python
drug_dict
```

```python
r = requests.post("https://arax.ncats.io/api/arax/v1.3/entity", json = drug_dict)
r.text
```

```python
r.json()
json_response = r.json()
```

```python
drugs_ids = []
for drug,response in json_response.items():
    drug_id_couple = [drug, response["id"]["identifier"]]
#     print(drug_id_couple)
    drugs_ids.append(drug_id_couple)   
drugs_ids      
```

```python
drugs_ids = pd.DataFrame(drugs_ids, columns=["drug", "normalized_drug_id"])
drugs_ids
```

```python
table = pd.merge(table, drugs_ids, left_on="Drug", right_on="drug", how="left")
table
```

```python
table = table.drop(['drug'], axis=1)
table
```

```python
genes = table["Gene"].tolist()
gene_dict = {
    "format": "minimal",
    "terms": genes
    
}
gene_dict
```

```python
r = requests.post("https://arax.ncats.io/api/arax/v1.3/entity", json = gene_dict)
json_response = r.json()
json_response
```

```python
gene_ids = []
for gene,response in json_response.items():
    gene_id_couple = [gene, response["id"]["identifier"]]
#     print(drug_id_couple)
    gene_ids.append(gene_id_couple)   
gene_ids  
```

```python
gene_ids = pd.DataFrame(gene_ids, columns=["gene", "normalized_gene_id"])
gene_ids
```

```python
table = pd.merge(table, gene_ids, left_on="Gene", right_on="gene", how="left")
table
```

```python
table = table.drop(['gene'], axis=1)
table
```

```python
table.to_csv("data.tsv", sep='\t', index=False)
```
